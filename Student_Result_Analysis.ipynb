```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/drive/MyDrive/student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
#gender distribution
plt.figure(figsize= (4,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#From the above chart we are analysed that:
#the number of females in the data is greater than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean", "ReadingScore":"mean", "WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores.
```

```python
sns.boxplot(data = df, y = "MathScore")
plt.show()
```

```python
sns.boxplot(data = df, y = "ReadingScore")
plt.show()
```

```python
#From this boxplot we are analysed that Math is comparitivily difficult subject instead of Reading and Writing to scores good marks.
```

```python
print(df["EthnicGroup"].unique())
```

```python
#Distribution of Eithic groups
groupA =df.loc[(df['EthnicGroup'] == "group A")].count()
groupB =df.loc[(df['EthnicGroup'] == "group B")].count()
groupC =df.loc[(df['EthnicGroup'] == "group C")].count()
groupD =df.loc[(df['EthnicGroup'] == "group D")].count()
groupE =df.loc[(df['EthnicGroup'] == "group E")].count()

lab = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist, labels =lab, autopct= "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
